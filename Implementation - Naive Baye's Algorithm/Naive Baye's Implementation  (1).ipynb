```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
x,y=load_iris(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

### 0 is indicating as male and 1 is indicating as a female

```python
tips['sex']=label.fit_transform(tips['sex'])
```

### 0 is indicating as No and 1 is indicating as Yes

```python
tips['smoker']=label.fit_transform(tips['smoker'])
```

```python
tips.head()
```

```python
tips.tail()
```

```python
tips.shape
```

```python
import pandas as pd
```

```python
tips=pd.get_dummies(tips, columns=['day', 'time'], drop_first=True)
```

```python
tips.head()
```

```python
x=tips.drop(columns='smoker')
y=tips['smoker']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
gaussian=GaussianNB()
```

```python
gaussian.fit(x_train,y_train)
```

```python
y_pred=gaussian.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
